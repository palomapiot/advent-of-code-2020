--- Day 11: Seating System ---
    
Your plane lands with plenty of time to spare. The final leg of your journey is a ferry that goes directly to the tropical island where you can finally start your vacation. As you reach the waiting area to board the ferry, you realize you're so early, nobody else has even arrived yet!

By modeling the process people use to choose (or abandon) their seat in the waiting area, you're pretty sure you can predict the best place to sit. You make a quick map of the seat layout (your puzzle input).

The seat layout fits neatly on a grid. Each position is either floor (.), an empty seat (L), or an occupied seat (#). For example, the initial seat layout might look like this:

Now, you just need to model the people who will be arriving shortly. Fortunately, people are entirely predictable and always follow a simple set of rules. All decisions are based on the number of occupied seats adjacent to a given seat (one of the eight positions immediately up, down, left, right, or diagonal from the seat). The following rules are applied to every seat simultaneously:

If a seat is empty (L) and there are no occupied seats adjacent to it, the seat becomes occupied.
If a seat is occupied (#) and four or more seats adjacent to it are also occupied, the seat becomes empty.
Otherwise, the seat's state does not change.
Floor (.) never changes; seats don't move, and nobody sits on the floor.

```python
import copy
```

```python
with open('day11.txt', 'r') as file:
    data = [list(x.strip('\n')) for x in file]
```

```python
max_pos = len(data) - 1
```

```python
def seating_system(seats):
    updated_seats = copy.deepcopy(seats)
    for i, i_val in enumerate(seats):
        for j, j_val in enumerate(i_val):
            if j_val == 'L' and adj_occupied_seats(i, j, seats) == 0:
                updated_seats[i][j] = '#'
            elif j_val == '#' and adj_occupied_seats(i, j, seats) >= 4:
                updated_seats[i][j] = 'L'
            
    if updated_seats == seats:
        return seats
    else:
        return seating_system(updated_seats)
```

```python
def adj_occupied_seats(i, j, seats):
    if i == 0:
        if j == 0:
            return is_occupied(i, j+1, seats) + is_occupied(i+1, j+1, seats) + is_occupied(i+1, j, seats)
        elif j == max_pos:
            return is_occupied(i, j-1, seats) + is_occupied(i+1, j, seats) + is_occupied(i+1, j-1, seats)
        else:
            return is_occupied(i, j-1, seats) + is_occupied(i, j+1, seats) + is_occupied(i+1, j-1, seats)+ is_occupied(i+1, j, seats)+ is_occupied(i+1, j+1, seats)
    
    if i == max_pos:
        if j == 0:
            return is_occupied(i, j+1, seats) + is_occupied(i-1, j+1, seats) + is_occupied(i-1, j, seats)
        elif j == max_pos:
            return is_occupied(i, j-1, seats) + is_occupied(i-1, j, seats) + is_occupied(i-1, j-1, seats)
        else:
            return is_occupied(i, j-1, seats) + is_occupied(i, j+1, seats) + is_occupied(i-1, j-1, seats)+ is_occupied(i-1, j, seats) + is_occupied(i-1, j+1, seats)
    
    if j == 0:
        return is_occupied(i-1, j, seats) + is_occupied(i-1, j+1, seats) + is_occupied(i, j+1, seats)+ is_occupied(i+1, j, seats) + is_occupied(i+1, j+1, seats)
    if j == max_pos:
        return is_occupied(i-1, j, seats) + is_occupied(i-1, j-1, seats) + is_occupied(i, j-1, seats)+ is_occupied(i+1, j, seats) + is_occupied(i+1, j-1, seats)
    
    return is_occupied(i-1, j, seats) + is_occupied(i-1, j+1, seats) + is_occupied(i-1, j-1, seats) + is_occupied(i, j+1, seats) + is_occupied(i, j-1, seats) + is_occupied(i+1, j, seats) + is_occupied(i+1, j-1, seats) + is_occupied(i+1, j+1, seats)    
```

```python
def is_occupied(i, j, seats):
    if seats[i][j] == '#':
        return 1
    else:
        return 0
```

```python
seats = seating_system(data)
```

```python
total_occ = 0
for l in seats:
    for c in l:
        if c == '#':
            total_occ += 1
print(total_occ) # 2470
```

```python
def seating_system_2(seats):
    updated_seats = copy.deepcopy(seats)
    for i, i_val in enumerate(seats):
        for j, j_val in enumerate(i_val):
            if j_val == 'L' and adj_occupied_seats_2(i, j, seats) == 0:
                updated_seats[i][j] = '#'
            elif j_val == '#' and adj_occupied_seats_2(i, j, seats) >= 5:
                updated_seats[i][j] = 'L'
            
    if updated_seats == seats:
        return seats
    else:
        return seating_system_2(updated_seats)
```

```python
def adj_occupied_seats_2(i, j, seats):
    if i == 0:
        if j == 0:
            return is_occupied(i, j+1, seats, 'RIGHT') + is_occupied(i+1, j+1, seats, 'DOWN-RIGHT') + is_occupied(i+1, j, seats, 'DOWN')
        elif j == max_pos:
            return is_occupied(i, j-1, seats, 'LEFT') + is_occupied(i+1, j, seats, 'DOWN') + is_occupied(i+1, j-1, seats, 'DOWN-LEFT')
        else:
            return is_occupied(i, j-1, seats, 'LEFT') + is_occupied(i, j+1, seats, 'RIGHT') + is_occupied(i+1, j-1, seats, 'DOWN-LEFT')+ is_occupied(i+1, j, seats, 'DOWN')+ is_occupied(i+1, j+1, seats, 'DOWN-RIGHT')
    
    if i == max_pos:
        if j == 0:
            return is_occupied(i, j+1, seats, 'RIGHT') + is_occupied(i-1, j+1, seats, 'UP-RIGHT') + is_occupied(i-1, j, seats, 'UP')
        elif j == max_pos:
            return is_occupied(i, j-1, seats, 'LEFT') + is_occupied(i-1, j, seats, 'UP') + is_occupied(i-1, j-1, seats, 'UP-LEFT')
        else:
            return is_occupied(i, j-1, seats, 'LEFT') + is_occupied(i, j+1, seats, 'RIGHT') + is_occupied(i-1, j-1, seats, 'UP-LEFT')+ is_occupied(i-1, j, seats, 'UP') + is_occupied(i-1, j+1, seats, 'UP-RIGHT')
    
    if j == 0:
        return is_occupied(i-1, j, seats, 'UP') + is_occupied(i-1, j+1, seats, 'UP-RIGHT') + is_occupied(i, j+1, seats, 'RIGHT')+ is_occupied(i+1, j, seats, 'DOWN') + is_occupied(i+1, j+1, seats, 'DOWN-RIGHT')
    if j == max_pos:
        return is_occupied(i-1, j, seats, 'UP') + is_occupied(i-1, j-1, seats, 'UP-LEFT') + is_occupied(i, j-1, seats, 'LEFT')+ is_occupied(i+1, j, seats, 'DOWN') + is_occupied(i+1, j-1, seats, 'DOWN-LEFT')
    
    return is_occupied(i-1, j, seats, 'UP') + is_occupied(i-1, j+1, seats, 'UP-RIGHT') + is_occupied(i-1, j-1, seats, 'UP-LEFT') + is_occupied(i, j+1, seats, 'RIGHT') + is_occupied(i, j-1, seats, 'LEFT') + is_occupied(i+1, j, seats, 'DOWN') + is_occupied(i+1, j-1, seats, 'DOWN-LEFT') + is_occupied(i+1, j+1, seats, 'DOWN-RIGHT')    
```

```python
def is_occupied(i, j, seats, direction):
    if direction == 'RIGHT':
        return is_occupied_aux(i, j, seats, '', '+')
    if direction == 'LEFT':
        return is_occupied_aux(i, j, seats, '', '-')
    if direction == 'DOWN':
        return is_occupied_aux(i, j, seats, '+', '')
    if direction == 'UP':
        return is_occupied_aux(i, j, seats, '-', '')
    
    if direction == 'UP-RIGHT':
        return is_occupied_aux(i, j, seats, '-', '+')
    if direction == 'UP-LEFT':
        return is_occupied_aux(i, j, seats, '-', '-')
    if direction == 'DOWN-RIGHT':
        return is_occupied_aux(i, j, seats, '+', '+')
    if direction == 'DOWN-LEFT':
        return is_occupied_aux(i, j, seats, '+', '-')
```

```python
def is_occupied_aux(i, j, seats, op_i, op_j):
    while i >= 0 and i <= max_pos and j >= 0 and j <= max_pos:
        if seats[i][j] == '#':
            return 1
        elif seats[i][j] == 'L':
            return 0
        
        if op_i == '+':
            i += 1
        elif op_i == '-':
            i -= 1
        if op_j == '+':
            j += 1
        elif op_j == '-':
            j -= 1
    return 0
```
